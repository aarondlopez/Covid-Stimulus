```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
data = [["Direct Payments", 0, 445000000000],
        ["Tax Incentives", 212500000000, 390000000000],
        ["Expanded Unemployment", 47500000000, 561000000000],
        ["Small Business Loans", 730000000000, 400000000000],
        ["Public Health", 128000000000, 483000000000],
        ["Infrastructure", 85000000000, 0],
        ["Alternative Energy", 22000000000, 0],
        ["Education", 109000000000, 0],
        ["Research", 18000000000, 0],
        ["State and Local Government", 0, 150000000000],
        ["Emergency Lending", 0, 510000000000],
        ["Other*", 0, 541000000000]]

df = pd.DataFrame(data, columns=["Category", "ARRA", "CARES+"])
```

```python
tidy = df.melt(id_vars='Category').rename(columns=str.title)
tidy.head()
```

```python
tidy = tidy.rename(columns={"Variable":"Stimulus Package", "Value":"Dollars"})
```

```python
sns.set_style("darkgrid")
plt.style.use('fivethirtyeight')
fig, ax1 = plt.subplots(figsize=(10, 7))
x_axis = tidy["Dollars"]/1000000000
sns.barplot(x=x_axis, y="Category", hue="Stimulus Package", data=tidy, ax=ax1)
ax1.set(xlabel="Billion (USD)", ylabel="", title="2009 Financial Crises vs. Coronavirus Stimulus Spending")
ax1.text(0, 13.5, "*Transportation, food stamps, childcare assistance, rental assistance, and other programs")
```
